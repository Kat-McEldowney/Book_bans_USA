```python
import pandas as pd
```

```python
# read the CSVs
index_1 = 'PEN_Ams_Index_07_01_21_to_06_30_22.csv'
index_2 = 'PEN_Ams_Index_07_01_22_to_06_30_23.csv'

# Read CSV files into pandas DataFrames
df1 = pd.read_csv(index_1, skiprows =2)
df1.head()
```

```python

columns = ['Title', 'Author', 'Secondary Author', 'Illustrators(s)', 'Translatror(s)', 
           'Series Name', 'State', 'District', 'Date of Challenge', 'Ban Status', 'Origin of Challenge']
df2 = pd.read_csv(index_2, header=None)
df2.columns = columns

df2.head()
```

```python
# Concatenate the DataFrames
combined_df = pd.concat([df1, df2], ignore_index=True)

# Display the concatenated DataFrame
combined_df.head()
```
